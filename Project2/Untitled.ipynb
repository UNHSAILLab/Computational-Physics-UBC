```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def exp(tList):
    
    return np.cos(tList), np.sin(tList)
```

```python
tList = np.arange(0,5,0.005)
N = tList.shape[0]
xList, yList = exp(tList)
```

```python
plt.plot(xList, yList)
```

```python
import numpy as np
%matplotlib notebook
from matplotlib import pyplot as plt
from matplotlib.animation import FuncAnimation
plt.style.use('seaborn-pastel')


fig = plt.figure()
ax = plt.axes(xlim=(0, 4), ylim=(-2, 2))
line, = ax.plot([], [], 'o', lw=3)

def init():
    line.set_data([], [])
    ax.set_xlim([-1,1])
    ax.set_ylim([-1,1])
    return line,
def animate(i):
    ydata = yList[:i]
    xdata = xList[:i]
    line.set_data(xdata, ydata)
    ax.set_title("frame: {}".format(i))
    return line,

anim = FuncAnimation(fig, animate, init_func=init,
                               frames=N, interval=5, blit=True)
```

```python
def func(xList,p):
    x0 = 0.5
    delta = 0.2
    return np.cos(p*xList)*np.exp(-((xList-x0)/delta)**2),np.sin(p*xList)*np.exp(-((xList-x0)/delta)**2)
```

```python
xList = np.linspace(0,1,100*3)
X,Y = func(xList,np.pi*2**(6))
```

```python
plt.scatter(X,Y, c = xList, cmap="hsv",lw=0.1)
plt.colorbar()
```

```python
Norm = (X**2 + Y**2)**0.5
```
