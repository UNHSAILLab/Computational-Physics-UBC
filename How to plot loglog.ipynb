```python
import numpy as np
import matplotlib.pyplot as plt
```

# In nutshell:

### y = $x^{something}$ --> use loglog
### y = $\log(x)$ --> use xlog
### y = $something^{x}$ --> use ylog

# Method
On paper, we should do the following change of variables:

$x = 10^X$

$y = 10^Y$

# Example:  loglog plot

## This can be used to plot x^(something) in a linear form

$y= \sqrt{x}$

$y= x^0.5$

change of variable: $ x = 10^X$
change of variable: $y = 10^Y$

so we will have:

$10^Y = 10^(X/2)$
$\leftarrow Y = X/2$

```python
x = np.arange(1,10, 0.01)
y = np.sqrt(x)
plt.plot(x,y)
plt.title("regular plot")
plt.xlabel("x")
plt.ylabel("y")
```

```python
plt.plot(np.log10(x), np.log10(y))
plt.xlabel("X = log(x)")
plt.ylabel("Y = log(y)")
plt.title("loglog plot")
```

# Example: SemiLog plot (x log) $\rightarrow y= ln(x)$

## This can be used to plot the slowly growing functions (like ln) in a linear fashion

change of variable: $x = 10^X$

$\rightarrow y = \ln(10^X) = X . \ln(10)$

```python
x = np.arange(1,10,0.01)
y = np.log(x)

plt.plot(x,y)
plt.title("regular plot")
plt.xlabel("x")
plt.ylabel("y")
```

```python
plt.plot(np.log(x),y)
plt.title("xLog plot")
plt.xlabel("X = log(x)")
plt.ylabel("y")
```

# Example: SemiLog plot (y log) $\rightarrow y= \exp(x)$

## This can be used to plot the fast growing functions (like ln) in a linear fashion

change of variable: $y = 10^Y$

$\rightarrow 10^Y = exp(x) \rightarrow Y = x\log(e)$

```python
x = np.arange(1,10,0.01)
y = np.exp(x)

plt.plot(x,y)
plt.title("regular plot")
plt.xlabel("x")
plt.ylabel("y")
```

```python
plt.plot(x,np.log(y))
plt.title("yLog plot")
plt.xlabel("X = log(x)")
plt.ylabel("y")
```
