```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import operator as op
from functools import reduce

def ncr(n, r):
    r = min(r, n-r)
    numer = reduce(op.mul, range(n, n-r, -1), 1)
    denom = reduce(op.mul, range(1, r+1), 1)
    return numer // denom  # or / in Python 2


### I got this function online!
```

```python
N_A = 10
N_B = 20

Q = 200
OmegaList = np.zeros(Q)
Omega_AList = np.zeros(Q)
Omega_BList = np.zeros(Q)

for q_a in range(Q):
    Omega_A = ncr(N_A+q_a-1,q_a)
    Omega_B = ncr(N_B+(Q-q_a)-1, Q-q_a)
    Omega_total =  Omega_A*Omega_B 
    Omega_AList[q_a] = Omega_A
    Omega_BList[q_a] = Omega_B
    OmegaList[q_a] = Omega_total
```

```python
plt.plot(np.log(OmegaList))
plt.plot(np.log(Omega_AList))
plt.plot(np.log(Omega_BList))
plt.grid()
plt.xlabel("q_a")
```

```python
S_A = np.log(Omega_AList)
DS_A = -(S_A - np.roll(S_A,-1))[:-1]
```

```python
S_B = np.log(Omega_BList)[::-1]
DS_B = -(S_B - np.roll(S_B,-1))[:-1]
```

```python
S_total = np.log(OmegaList)
DS = -(S_total - np.roll(S_total,-1))
```

```python
plt.plot(1/DS_A)
plt.plot(1/DS_B[::-1])
plt.plot(1/DS)
plt.ylim([0,50])
plt.grid()
```
