```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
mpl.rc('font',size=14)
mpl.rc("figure", figsize=(8,4))
```

```python
df = pd.read_csv('alpha-range.csv')
```

```python
df
```

```python
df.columns[1:]
```

```python
plt.figure(figsize=(12,6))
for channel in df.columns[1:]: 
    plt.plot(df[channel],".",ms=3,label=channel+"kPa")
    idmax = df[channel].idxmax()
    maxValue = df[channel][idmax]
    plt.scatter(idmax, maxValue)
plt.title("Data")
plt.xlabel("Channel (Energy)")
plt.ylabel("Count")
plt.legend()
```
