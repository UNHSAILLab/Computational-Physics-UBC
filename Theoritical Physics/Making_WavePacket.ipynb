## Making WavePackets

wavepackets are very interesting for me. I have always been looking to find its functional form but after a while I found out that they can be expressed in the form of integrals

$$ P(x) = \int e^{\frac{(k-k_0)^2}{\sigma^2}} e^{i(kx)} dk $$

this integral means if we different sine waves with frequencies which has a guassian distribution around k_0 then the result will be a wave packet.

```python
dk = 0.05
kList = np.arange(-10,10,dk)
```

```python
dx = 0.05
xList = np.arange(-10,10,dx)
```

```python
X,K = np.meshgrid(xList, kList)
```

```python
k0 = 5
sigma = 1
```

```python
F1 = np.real(np.exp(-(sigma)**2 *(K-k0)**2)*np.exp(1j*(K*X)))
```

```python
k0 = 6
```

```python
plt.figure(figsize=(10,8))
plt.imshow(F1, extent=[-10,10,5,-5])
```

```python
integral1 = (0.5*(F1+np.roll(F1,-1, axis=0))[:-1,:]).sum(axis=0)
```

```python
plt.plot(xList, integral1)
```

## Trying other forms:

Lets try and see how the following integral will look like:

$$ P(x) = \int e^{\frac{(k-k_0)^2}{\sigma^2}} e^{i(k-k0)x} dk $$

This form is important for me because in studying the dispersion of waves, we see this form of integral. At first glance one might say that it will also be wave packet. It is a wave packet but it will have a constant shape no matter what is the value of k_0. In fact it will be very similar to a wave packet with k_0 = 0. We can show this by plotting it, but also we can shwo this by a simple change of variable:

$$ ζ = k - k_0 $$

So the integral will be like the following:

$$ P(x) = \int e^{\frac{\zeta^2}{\sigma^2}} e^{i\zeta x} dk $$


```python
F2 = np.real(np.exp(-1*(K-k0)**2)*np.exp(1j*((K-k0)*X)))
```

```python
plt.figure(figsize=(10,8))
plt.imshow(F2, extent=[-10,10,5,-5])
```

```python
integral2 = (0.5*(F2+np.roll(F2,-1, axis=0))[:-1,:]).sum(axis=0)
plt.plot(integral2)
```
