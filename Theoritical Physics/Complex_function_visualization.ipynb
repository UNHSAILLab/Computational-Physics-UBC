```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon
```

```python
def meshPlotter(X,Y):
    plt.figure(figsize=(15,15))
    # xThreshold = 5
    # yThreshold = 5
    # X[Y>xThreshold] = np.NaN
    # Y[X>yThreshold] = np.NaN
    _ = plt.plot(X.T, np.imag(Y).T, 'k', alpha=0.4)
    _ = plt.plot(X, np.imag(Y), 'k', alpha=0.4)
```

```python
def f(z):
    
    return (z-1)/(z**2+z+1)
    # return np.log(z+10)**3


def saw(z):
    return np.ceil(z) - (z)


def angle(z):
    theta = np.angle(z)
    return theta + np.heaviside(-theta,0)*2*np.pi
```

```python
x = np.linspace(-0.65,0.65,50)
X,Y = np.meshgrid(x, x*1j)
```

```python
result = f(X+Y)
```

```python
import numpy as np

import matplotlib.pyplot as plt
import matplotlib
from matplotlib.patches import Polygon
from matplotlib.collections import PatchCollection

fig, ax = plt.subplots()
patches = []
num_polygons = 5
num_sides = 5

for i in range(num_polygons):
    polygon = Polygon(np.random.rand(num_sides ,2), True)
    patches.append(polygon)

p = PatchCollection(patches, cmap=matplotlib.cm.jet, alpha=0.4)

colors = 100*np.random.rand(len(patches))
p.set_array(np.array(colors))

ax.add_collection(p)

plt.show()
```
