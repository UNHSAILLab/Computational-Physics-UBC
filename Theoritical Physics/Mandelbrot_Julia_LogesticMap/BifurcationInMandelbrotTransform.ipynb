```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def generate(maxIter,saver):
    c = np.arange(-2,-1,0.001)
    z = np.zeros(c.shape)
    X = np.ones((c.shape[0],saver))*0.5
    for i in range(maxIter):
        z = z**2 + c
        if (maxIter-i<saver):
            X[:,saver-(maxIter-i)] = z
    return X,c
```

```python
maxIter = 20
saver = 10
X,c = generate(maxIter,saver)
```

```python
plt.figure(figsize=(12,8))
a = plt.plot(c,X[:,1:])
```
