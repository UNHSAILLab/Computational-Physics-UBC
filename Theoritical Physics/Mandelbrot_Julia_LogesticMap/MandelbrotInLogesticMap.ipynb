## Mandelbrot inside the Logestic Map

$x_{n+1} = r * x_n * (1-x_n)$

Here in this code, I generalize the concept of logistic map.
Here I allow the born rate (r) to be imaginary numbers as well.


```python
color = np.zeros(R.shape)
maxIter = 50
z = 0.5
for i,row in enumerate(R):
    for j,r in enumerate(row):
        n=0
        z= 0.5
        while np.abs(z)<=150 and n<maxIter:
            z = r * z * (1-z)
            n+=1
            
        color[i,j] = n
        
```

```python
d = 6
xstart = -2
x = np.linspace(xstart,xstart+d,1000)
y = np.linspace(-d/4,d/4,1000)*1j
XX,YY = np.meshgrid(x,y)
R = XX+YY
```

```python
maxIter = 30
thresh = 2
def mandel(R,maxIter,thresh):
    color = np.zeros(R.shape)+maxIter
    Z = np.ones(color.shape)*0.5
    for i in range(maxIter):
        Z = R*Z*(1-Z)
        Zabs = np.abs(Z)
        color[Zabs>=thresh]=i
        Z[Zabs>=thresh]=float('nan')
        
    return color
```

```python
color = mandel(R,maxIter, thresh)
color = maxIter - color
```

```python
fig = plt.figure(figsize=(18,10))
plt.imshow(color,aspect='auto',cmap='turbo',extent=[xstart,xstart+d,-d/4,d/4])
plt.colorbar()
```

```python
def julia(Z,maxIter,thresh):
    color = np.zeros(Z.shape)+maxIter
    R = np.zeros(color.shape)+0.57j+0.85
    for i in range(maxIter):
        Z = R*Z*(1-Z)
        Zabs = np.abs(Z)
        color[Zabs>=thresh]=i
        Z[Zabs>=thresh]=float('nan')
        
    return color
```

```python
Z = XX+YY
color = julia(Z,maxIter, thresh)
color = maxIter-color
```

```python
plt.figure(figsize=(18,10))
plt.imshow(color,aspect='auto',cmap='turbo')
plt.colorbar()
```
