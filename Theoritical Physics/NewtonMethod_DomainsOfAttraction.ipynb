```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5,5, 1500)
```

```python
X,Y = np.meshgrid(x,x*1j)
```

```python
def trans(Z):
    # Z = X+Y
    return -(Z**5-1)/(5*Z**4) + Z
```

```python
colors = np.zeros(X.shape)
Z = X+Y
for i in range(20):

    Z = trans(Z)
```
