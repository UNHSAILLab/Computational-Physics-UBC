```python
import numpy as np
import matplotlib.pyplot as plt
```

Euler method to solve the following differential equation

$\ddot \theta = -\theta$

First Step: writing the second order ODE in form of two first order ODEs

$\dot \theta = \omega$

$\dot \omega = - \theta$

Eurler Method:

$\theta_{i+1} = \theta_{i} + \omega _{i} \Delta t$

$\omega_{i+1} = \omega_i + a_i \Delta t$

$t_{i+1} = t_{i} + \Delta t$

in which $a_i$ is acceleration $\ddot \theta = -\theta$


This method is not a good way to solve ODEs (specially those that correspond to oscillatory motion). Because if we wirte the enegy a pendulum that we want to solve it using the Euler method, we will have:

$E = \dfrac{1}{2} m l^2 \omega^2 + mgl(1-cos(\theta))$

and if we replace the values of $\theta_{i+1}, \omega{i+1}$ to the equation above, we will have:


$E_{i+1} = E_{i} + \dfrac{1}{2}mgl(\omega_{i}^2 + \dfrac{g}{l} \theta_i ^2)(\Delta t)^2$

Since the second term is always positive, the value of E will increase at each iteration, no matter what value we choose for $\Delta t$. Indeed, smaller values of $\Delta t$ will lead to slower divergence, but divergence is always there.

One way to solve this problem in Euler-Cromer Method


### Euler Method

```python
class Env:
    def __init__(self, lengthD, timeD, massD):
        self.lengthD = lengthD
        self.timeD = timeD
        self.massD = massD
        
        
    def Sim2SI(self, value, quantity):
        if quantity == "mass":
            return value*self.massD
        if quantity == "length":
            return value*self.lengthD
        if quantity == "time":
            return value*self.timeD
        
    def SI2Sim(self, value, quantity):
        return value / self.Sim2SI(1, quantity)
    
```

```python
class Pendulum:
    def __init__(self, position, velocity, dt):
        self.position = position
        self.velocity = velocity
        self.nextVelocity = velocity
        self.acceleration = -self.position
        self.position_ = self.position
        self.dt = dt
        
    def move(self):
        '''
        theta(i+1) = theta(i) + w(i)*dt
        w(i+1) = w(i) + a(i)*dt
        
        '''
        self.position_ = self.position
        self.position = self.updateVariable(self.position, self.velocity)
        self.acceleration = -self.position_
        self.velocity = self.updateVariable(self.velocity, self.acceleration)
        
    def updateVariable(self, variable, ddtVariable):
        newValue = self.EulerSolver(variable, ddtVariable, self.dt)
        return newValue
        
    def EulerSolver(self, f0, derivitive, dh):
        return (f0 + derivitive*dh)
    
    def moveForTimeT(self, time):
        N = time/self.dt
        positionList = np.zeros(int(N))
        EnergyList = np.zeros(int(N))
        velocityList = np.zeros(int(N))
        timeList = np.zeros(int(N))
        t = 0
        print(self.position)
        
        
        for i in range(int(N)):
            self.move()
            t += dt
            timeList[i] = (t)
            positionList[i] = self.position  
            EnergyList[i] = self.calculateEnergy()
            velocityList[i] = self.velocity
            
            
        return positionList, EnergyList, velocityList, timeList
    
    def calculateEnergy(self):
        return (0.5*self.velocity**2 )
        
        
```

```python
g = 10
L = 1
t_max = 10
dt_real = 0.04
myEnv = Env(L,(0.1)**0.5,1) #length, time, mass
```

```python
myEnv.Sim2SI(1, 'time')
```

```python
position = 0.1
velocity = 0
dt = myEnv.SI2Sim(dt_real, 'time')
```

```python
dt
```

```python
tmax_simulation =  myEnv.SI2Sim(t_max, 'time')
pendulum = Pendulum(position, velocity, dt)
```

```python
positionList, energyList, velocityList, timeList = pendulum.moveForTimeT(tmax_simulation)
```

```python
timeList = myEnv.Sim2SI(timeList, 'time')
```

```python
fig = plt.figure(figsize = (10,10))
fig.add_subplot(2,2,1)
plt.plot(timeList,positionList)
fig.add_subplot(2,2,2)
plt.plot(timeList,velocityList)
fig.add_subplot(2,2,3)
plt.plot(timeList,0.5*velocityList**2 + 1 - np.cos(positionList))
#plt.ylim([0,0.0001])
fig.add_subplot(2,2,4)
plt.plot(positionList, velocityList)
```

### Euler Cromer Method


$\omega_{i+1} = \omega_i + a_i \Delta t$

$\theta_{i+1} = \theta_{i} + \omega _{i+1} \Delta t$


$t_{i+1} = t_{i} + \Delta t$

Euler Cromer method preserves the energy at each cycle. you can proof that using the same approach usesd in the previouse section to show that the Euler method is not stable

```python
class Pendulum:
    def __init__(self, position, velocity, dt):
        self.position = position
        self.velocity = velocity
        self.nextVelocity = velocity
        self.acceleration = -self.position
        self.position_ = self.position
        self.dt = dt
        
    def move(self):
       
        self.acceleration = -self.position
        self.velocity = self.updateVariable(self.velocity, self.acceleration)
        self.position = self.updateVariable(self.position, self.velocity)
        
        
    def updateVariable(self, variable, ddtVariable):
        newValue = self.EulerCrommerSolver(variable, ddtVariable, self.dt)
        return newValue
        
    def EulerCrommerSolver(self, f0, derivitive, dh):
        return (f0 + derivitive*dh)
    
    def moveForTimeT(self, time):
        N = time/self.dt
        positionList = np.zeros(int(N))
        EnergyList = np.zeros(int(N))
        velocityList = np.zeros(int(N))
        timeList = np.zeros(int(N))
        t = 0
        print(self.position)
        
        
        for i in range(int(N)):
            self.move()
            t += dt
            timeList[i] = (t)
            positionList[i] = self.position  
            EnergyList[i] = self.calculateEnergy()
            velocityList[i] = self.velocity
            
            
        return positionList, EnergyList, velocityList, timeList
    
    def calculateEnergy(self):
        return (0.5*self.velocity**2 )
        
```

```python
g = 10
L = 1
t_max = 10
dt_real = 0.04
myEnv = Env(L,(0.1)**0.5,1) #length, time, mass
```

```python
myEnv.Sim2SI(1, 'time')
```

```python
position = 0.1
velocity = 0
dt = myEnv.SI2Sim(dt_real, 'time')
```

```python
timeList = myEnv.Sim2SI(timeList, 'time')
```
