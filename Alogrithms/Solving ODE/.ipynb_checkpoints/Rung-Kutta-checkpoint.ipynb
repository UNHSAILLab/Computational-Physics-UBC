# Theory and Derivation

## Single Rung-Kutta

Here I am going to solve the following single first order differential equation

```python
def RK4(y0, f, x0, x_end, l):
    N = np.power(2,l)
    h = (x_end - x0)/N
    xList = np.arange(x0, x_end, h)
    yList = np.zeros(xList.shape, dtype= 'float')
    yList[0] = y0
    
    for i in range(xList.shape[0]-
1):
        x0 = xList[i]
        y0 = yList[i]
        f0 = f(x0,y0)
        f1 = f(x0+h/2, y0+h/2*f0)
        f2 = f(x0+h/2, y0+h/2*f1)
        f3 = f(x0+h, y0+h*f2)
        yList[i+1] = y0 + h/6*(f0+2*f1+2*f2+f3)
        
    return xList, yList
```

```python
def f(x,y):
    return -y
```

```python
t5,y5 = RK4(10,f,0,5,5)
t6,y6 = RK4(10,f,0,5,6)
t7,y7 = RK4(10,f,0,5,7)
```

```python
plt.plot(t5,(y6[::2]-y5))
plt.plot(t5,y7[::4]-y6[::2])
```

## Multiple Rung-Kutta


I will be soleving the system of first order differential equations

```python
def F(x,Y):
    F0 = Y[1]
    F1 = -50*Y[0]-x*Y[1]
    return np.array([F0,F1])
```

```python
def VectorRK4(Y0,F,x0,x_end,l):
    N = np.power(2,l)
    h = (x_end - x0)/N
    xList = np.arange(x0, x_end, h)
    yList = np.zeros((xList.shape[0], Y0.shape[0]), dtype='float')
    yList[0,:] = Y0
    
    for i in range(xList.shape[0]-1):
        x0 = xList[i]
        Y0 = yList[i,:]
        F0 = F(x0,Y0)
        F1 = F(x0+h/2, Y0+h/2*F0)
        F2 = F(x0+h/2, Y0+h/2*F1)
        F3 = F(x0+h, Y0+h*F2)
        yList[i+1,:] = Y0 + h/6*(F0+2*F1+2*F2+F3)
        
    return xList, yList
    
```

```python
Y0 = np.array([1,-2])
x,Y = VectorRK4(Y0,F,0,20,10)
```

```python
plt.plot(x,Y[:,0])
```

```python
plt.plot(x,Y[:,1])
```

## More Structured RK4

```python
def VectorRK4(Y0,F,x0,x_end,l):
    N = np.power(2,l)
    h = (x_end - x0)/N
    xList = np.arange(x0, x_end, h)
    yList = np.zeros((xList.shape[0], Y0.shape[0]), dtype='float')
    yList[0,:] = Y0
    
    for i in range(xList.shape[0]-1):
        x0 = xList[i]
        Y0 = yList[i,:]
        F0 = F(x0,Y0)
        F1 = F(x0+h/2, Y0+h/2*F0)
        F2 = F(x0+h/2, Y0+h/2*F1)
        F3 = F(x0+h, Y0+h*F2)
        yList[i+1,:] = Y0 + h/6*(F0+2*F1+2*F2+F3)
        
    return xList, yList
    
```

```python
def F(x,Y):
    F = np.zeros(Y.shape[0], dtype='float')
    for i in range(F.shape[0]-1):
        F[i] = Y[i+1]
    
    F[-1] = 1*(1-Y[0]**2)*Y[1] - Y[0]
    return F
```

## Van der Pol Oscillator

```python
def VectorRK4(mu,Y0,F,x0,x_end,l):
    N = np.power(2,l)
    h = (x_end - x0)/N
    xList = np.arange(x0, x_end, h)
    yList = np.zeros((xList.shape[0], Y0.shape[0]), dtype='float')
    yList[0,:] = Y0
    
    for i in range(xList.shape[0]-1):
        x0 = xList[i]
        Y0 = yList[i,:]
        F0 = F(mu,x0,Y0)
        F1 = F(mu,x0+h/2, Y0+h/2*F0)
        F2 = F(mu,x0+h/2, Y0+h/2*F1)
        F3 = F(mu,x0+h, Y0+h*F2)
        yList[i+1,:] = Y0 + h/6*(F0+2*F1+2*F2+F3)
        
    return xList, yList
```

```python
def F(mu,x,Y):
    F = np.zeros(Y.shape[0], dtype='float')
    for i in range(F.shape[0]-1):
        F[i] = Y[i+1]
    
    F[-1] = mu*(1-Y[0]**2)*Y[1] - Y[0]
    return F
```

```python
Y0 = np.array([0,4])
mu = 0.1

x,Y = VectorRK4(mu,Y0,F,0,100,10)

plt.figure(figsize=(15,5))
plt.plot(x,Y[:,0])
```

```python
mu = 0.5
YList = []

for i in range(50):
    Y0 = np.array(np.random.random(2)*6-3)
    x,Y = VectorRK4(mu,Y0,F,0,100,10)
    YList.append(Y)
```

```python
plt.figure(figsize=(12,8))
for Y in YList:
    plt.plot(Y[:,0], Y[:,1],c='k')
```

```python
muList = np.linspace(0.00,4,30)
YList = []
for mu in muList:
    Y0 = np.array([0.5,0.5])
    x,Y = VectorRK4(mu,Y0,F,0,100,12)
    YList.append(Y)
```
