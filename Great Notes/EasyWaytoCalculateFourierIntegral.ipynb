## Easy way to calculate the fourier integral

Here in this note book will be discussing a simple way to calculate the fourier integral.

The fourier integral is like this:

$$ F(x) =  \int \tilde F(k) e^{ikx} dk $$


Let's consider the function inside the integral:

$$ G(k,x) =  \tilde F(k) e^{ikx} $$

it is a 2D function. We can claculate its value using meshgrid trick in python.
So we will have a 2d matrix. depending on the axis that corresponds to the integral variable, we can integrate the matrix along that axis (using the np.roll function in numpy makes this step a one line code)

See the following example. Here I am integrating using the trapezoidal integration

```python
dk = 0.05
kList = np.arange(-10,10,dk)
```

```python
dx = 0.05
xList = np.arange(-10,10,dx)
```

```python
X,K = np.meshgrid(xList, kList)
```

# Example1: Wave packet

```python
def F(k):
    k0 = 3
    return np.exp(-(1)**2 *(K-k0)**2)
```

```python
G = F(K)*np.exp(1j*(K*X))
```

```python
plt.figure(figsize=(10,8))
plt.imshow(G.real, extent=[-10,10,5,-5])
```

```python
plt.figure(figsize=(10,8))
plt.imshow(G.imag, extent=[-10,10,5,-5])
```

```python
integral = (0.5*(G+np.roll(G,-1, axis=0))[:-1,:]).sum(axis=0)      ## Trapezoidal integraion
```

```python
plt.plot(xList, integral.real)
```
